```python
import json
```

```python
path = '/kaggle/input/ner-data/'
with open(path + 'train.json', 'r') as file:
  train = json.loads(file.read())

with open(path + 'test.json', 'r') as file:
  test = json.loads(file.read())
```

```python
import random
random.seed(23)
```

```python
ids = [doc['id'] for doc in train]
train_ids = random.sample(ids, k=int(len(train) * 0.8))
X = [doc for doc in train if doc['id'] in train_ids]
y = [doc for doc in train if doc['id'] not in train_ids]
print(len(train) == (len(X) + len(y)))
```

```python
import spacy
```

```python
model_path = '/kaggle/input/ner-model/model-best'
nlp_ner = spacy.load(model_path)
```

```python
def extract_entity(text, label):
    entity = {'extracted_part': {'text': [''], 'answer_start': [0], 'answer_end': [0]}}
    entities = nlp_ner(text)
    if entities:
        for ent in entities.ents:
            if ent.label_.lower() == label:
                entity['extracted_part']['text'] = [ent.text]
                entity['extracted_part']['answer_start'] = [ent.start_char]
                entity['extracted_part']['answer_end'] = [ent.end_char]
    return entity
```

```python
from tqdm import tqdm
```

```python
examples_count = len(y)
exact_answers = 0
correct_answers = 0
for doc in tqdm(y):
    true_label = doc['extracted_part']['text'][0]
    label = extract_entity(doc['text'], doc['label'])['extracted_part']['text'][0]
    print(true_label)
    print(label)
    if true_label == label:
        print('correct')
        exact_answers += 1
    elif (label in true_label or true_label in label) and true_label != '' and label != '':
        print('semi correct')
        correct_answers += 1
    else:
        print('not correct')
    print('--'*25)
```

```python
exact_answers / examples_count
```

```python
(correct_answers + exact_answers) / examples_count
```

```python
for i in tqdm(range(len(test))):
    test[i].update(extract_entity(test[i]['text'], test[i]['label']))
```

```python
for doc in test:
    if 'extracted_part' not in doc:
        print(doc)
```
